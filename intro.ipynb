```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Input
josb_assets_list = [101.783, 90.671, 81.41, 77.144, 82.515, 84.751, 88.954, 108.457, 133.993, 186.511, 231.83, 304.832, 368.392, 440.098, 491.366, 556.364, 662.037, 813.612, 894.847, 935.738]
mw_assets_list = [160.494, 204.105, 295.478, 379.415, 403.732, 611.195, 707.734, 717.869, 769.313, 869.198, 993.322, 1123.274, 1096.952, 1256.467, 1187.73, 1232.106, 1320.318, 1405.952, 1496.347, 1555.23]
year_list = [1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013]
```

```python
# Plot Graph

plt.figure(figsize=(28,16), dpi=80)
plt.plot(year_list,josb_assets_list,label="Joseph A Bank",linewidth=8)
plt.plot(year_list,mw_assets_list,'g-.',label="Men's Wearhouse",linewidth=8)
xtick_location = np.arange(1994,2013,2)
plt.xticks(ticks=xtick_location, rotation=0, fontsize=50, horizontalalignment='center')
plt.legend(bbox_to_anchor=(0.50, 0.95),fontsize=50)
plt.yticks(fontsize=50)
plt.grid(axis='both')
plt.xlabel('Year', fontsize=65)
plt.ylabel('Total Assets', fontsize=65)
```

```python
random_values = np.random.normal(1000,200,10000) # normal(mean_value, std, number_of_values)
bins = np.linspace(400,1600,30)
plt.figure(figsize=(24,16), dpi= 40)
plt.hist(random_values,bins)
plt.xlabel('Total Assets', fontsize = 65)
plt.ylabel('Frequency', fontsize = 65)
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.grid(axis='both')
plt.title('Histogram of Asset Values in NYSE', fontsize = 65)
```

```python
import scipy as sp
# Probability density function
sp.stats.norm.pdf(14,10,2)
```

```python
# Cummulative density function
sp.stats.norm.cdf(14,10,2)
```

```python
# Inverse tail distribution (Inverse survival function)
sp.stats.norm.isf(1-0.9772498680518208,10,2)
```

```python
# Inverse function (survival function)
sp.stats.norm.sf(14,10,2)
```

```python
# Percent point function (inverse of cdf  percentiles): standard normal distribution
# Considering a Confidence interval = 95%
# Z_upper = ppf(0.975,0,1) [0.25, 0.95, 0.25(Z_upper)]
# Z_lower = ppf(0.975,0,1) [0.25(Z_lower), 0.95, 0.25]

sp.stats.norm.ppf(0.9772498680518208,10,2)
```
